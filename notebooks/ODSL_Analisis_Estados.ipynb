```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
# La siguiente l√≠nea es para ver las imagenes dentro del notebook
%matplotlib inline
```

```python
df = pd.read_csv('penal_df.csv',delimiter=',')
```

```python
print('Dataset info: \n')
df.info()
```

```python
print('Dataset Describe: \n')
df.describe()
```

```python
print('Conteo de Estados: \n')
df.est_descr.value_counts()
```

```python
print('Correlaci√≥n entre valores del Dataset: \n')
df.corr()
```

```python
print('Correlaci√≥n entre valores del Dataset: \n')
df2=df.loc[:,['org_idradactual','est_id','oju_id','exp_anio','exp_fojas','exp_id']]
fecorr= df2.corr()
fecorr
```

```python
print('Gr√°fico Correlaci√≥n entre valores del Dataset: \n')

plt.rcParams['figure.figsize'] = (12,10)
# Seteamos opciones de pandas sobre las columnas y su ancho
pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)

sns.heatmap(fecorr, xticklabels=fecorr.columns, yticklabels=fecorr.columns);
```

```python
f= df.groupby('org_idradactual')['exp_id'].count().reset_index()
f.sort_values(by='exp_id',ascending=False)
print('Cantidad de Expedientes por juzgado: \n')
print(f)

print('\n M√©tricas de columna cantidad expedientes por juzgado: \n')
print(f.exp_id.describe())
```

```python
print('Gr√°fico de barras cantidad de expedientes por juzgado: \n')

plt.rcParams['figure.figsize'] = (12,8)

pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)

f.sort_values(by='exp_id',ascending=False,inplace=True)

a=sns.barplot(x='org_idradactual', y='exp_id',palette='pastel',data=f)
a.set_xlabel("Juzgado",fontsize=16)
a.set_ylabel("Cantidad de expedientes",fontsize=16)
a.set_title("Cantidad de Expedientes por Juzgado \n", fontsize=18)
plt.xticks(rotation=65);
```

```python
f_anio= df.groupby(['org_idradactual','exp_anio'])['exp_id'].count().reset_index()
f_anio.sort_values(by='exp_id',ascending=False,inplace=True)
print('Cantidad de Expedientes por juzgado y a√±o: \n')
print(f_anio)

print('\n M√©tricas de columna cantidad expedientes por juzgado y a√±o: \n')
print(f_anio.exp_id.describe())
```

```python
plt.rcParams['figure.figsize'] = (12,8)

pd.set_option('max_columns', 120)
pd.set_option('max_colwidth', 5000)


list=[2016,2017,2018,2019]
for i in list:
    
    print('Gr√°fico de barras cantidad de expedientes por juzgado, a√±o '+str(i)+': \n')
    f_anio_ =f_anio[f_anio.exp_anio==i].copy()
    sns.barplot(x='org_idradactual', y='exp_id',palette='pastel',data=f_anio_)
    plt.xticks(rotation=65)
    plt.show()

    
```

```python
f_estado= df.groupby(['org_idradactual','exp_anio','est_id'])['exp_id'].count().reset_index()
f_estado.sort_values(by='exp_id',ascending=False)
print('Cantidad de Expedientes por juzgado, estado y a√±o: \n')

print(f_estado)

f_estado_2016=f_estado[f_estado.exp_anio==2016].copy()
f_estado_2017=f_estado[f_estado.exp_anio==2017].copy()
f_estado_2018=f_estado[f_estado.exp_anio==2018].copy()
f_estado_2019=f_estado[f_estado.exp_anio==2019].copy()
```

```python
lab= np.array([ '1: Asignado',  '2: En Despacho',  '9: Pase', '10: En vista', '14: Archivado','16: Anulado', '46: Inicial', '62: Cerrado', '22: Migracion', '63: Principal', '52: Resuelto', '33: En Tramite', '21: Radicado', '76: Archivado', '20: Remitido por Inc',' 19: Prearchivo'])
```

```python
rows= f_estado['est_id'].unique()
columns= f_estado['org_idradactual'].unique()

df_estado_2016 = pd.DataFrame(columns=columns,index=rows)

# Iteraci√≥n por filas del DataFrame:
for indice, fila in f_estado_2016.iterrows():
    estado= f_estado_2016.est_id[indice]
    juzgado= f_estado_2016.org_idradactual[indice]
    exp= f_estado_2016.exp_id[indice]
    df_estado_2016.loc[estado,juzgado] = exp

df_estado_2016.fillna(value=0,inplace=True)

color_list = ['r', 'y', 'skyblue','bisque','green','peru','palegreen','violet','beige','orange','crimson','gold','steelblue','lavender','tan','aquamarine']

X_2016 = np.arange(df_estado_2016.shape[1])
labels_2016=rows

fig, ax = plt.subplots(figsize=(20, 8))
for i in range(df_estado_2016.shape[0]):
    plt.bar(X_2016, df_estado_2016.iloc[i], bottom = np.sum(np.array(df_estado_2016.iloc[0:i]), axis = 0), color = color_list[i % len(color_list)])

ax.set_title('Cantidad de Expedientes por Juzgado y Estado, A√±o 2016')
ax.set_ylabel('Cantidad de Expedientes')
ax.set_xlabel('Juzgado')
ax.set_xticks(X_2016)
plt.legend(labels=lab)
plt.show()
```
